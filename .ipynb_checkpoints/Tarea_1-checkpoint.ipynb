## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Director"

```python
%%sql
CREATE TABLE director (
    id_director INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INT NOT NULL
)
```

### Tabla "Actor"

```python
%%sql
CREATE TABLE actor (
    id_actor INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    nombre VARCHAR(45) NOT NULL,
    edad INT NOT NULL
)
```

### Tabla "Pelicula"

```python
%%sql
CREATE TABLE pelicula (
    id_pelicula INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    id_director INT NOT NULL,
    nombre VARCHAR(45) NOT NULL,
    genero VARCHAR(45) NOT NULL,
    duracion INT NOT NULL,
    fecha_publicacion DATE NOT NULL,
    FOREIGN KEY (id_director) REFERENCES director(id_director)
        ON DELETE RESTRICT
)
```

### Tabla "actor_participa_en_pelicula"

```python
%%sql
CREATE TABLE actor_participa_en_pelicula (
  id_actor INT,
  id_pelicula INT NOT NULL,  -- Add NOT NULL constraint
  FOREIGN KEY (id_actor) REFERENCES actor(id_actor),
  FOREIGN KEY (id_pelicula) REFERENCES pelicula(id_pelicula)
);
```

### Tabla  "Premio_mejor_director"

```python
%%sql
CREATE TABLE premio_mejor_director (
    id_premio INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,
    id_director INT NOT NULL,
    anio INT NOT NULL,
    FOREIGN KEY (id_director) REFERENCES director(id_director)
        ON DELETE CASCADE
        ON UPDATE RESTRICT
)
```

### Tabla  "Premio_mejor_actor"

```python
%%sql
CREATE TABLE premio_mejor_actor (
    id_premio INT GENERATED ALWAYS AS IDENTITY,
    id_actor INT NOT NULL,
    anio INT NOT NULL,
    FOREIGN KEY (id_actor) REFERENCES actor(id_actor)
        ON DELETE CASCADE
)
```

### Tabla  "Premio_mejor_pelicula"

```python
%%sql
CREATE TABLE premio_mejor_pelicula (
    id_premio INT GENERATED ALWAYS AS IDENTITY,
    id_pelicula INT NOT NULL,
    anio INT NOT NULL,
    FOREIGN KEY (id_pelicula) REFERENCES pelicula(id_pelicula)
        ON DELETE CASCADE
)
```

### Consulta 1
**Nombre de todas las películas publicadas desde el año pasado.**


**Ejemplo:**
|Nombre|
|------|
|Dune 2|
|...|


**Solución:** Se realiza una consulta en la que se selecciona el nombre de todas las películas (de la tabla "película"), siempre que se cumpla la condición de que su fecha de publicación (de la película) sea de hace un año o más reciente (siendo la fecha actual el límite). Para ello, se emplea la considicon WHERE en la que se indica que la fecha de publicación de toda película seleccionada en la consulta debe ser mayor o igual que la fecha actual (obtenida con "CURRENT_DATE") menos un año. De esta forma la consulta siempre estará actualizada a pesar de que los días pasen. "INTERVAL" se utiliza para realizar operaciones con fechas y tiempos, en este caso, la fecha actual se resta con "1 año".

```python
%%sql
SELECT nombre AS "Nombre"
FROM pelicula 
WHERE fecha_publicacion >= CURRENT_DATE - INTERVAL '1 year'; 
```

### Consulta 2
**Nombre y fecha de publicación de todas las películas del género drama que hayan sido publicadas en el año 2023.**


**Ejemplo:**

|Nombre     |Fecha de publicación|
|-----------|--------------------|
|Oppenheimer|21-7-2023          |
|...        |...                 |


**Solución:** _Seleccionamos los Nombres y fechas de publicacion que se pide que se impriman, y buscamos en la tabla pelicula todas las que tengan el genero drama, y que hayan sido publicadas en 2023_

```python
%%sql
SELECT nombre AS "Nombre", fecha_publicacion AS "Fecha de publicación"
FROM pelicula
WHERE genero = 'Drama'
AND EXTRACT(YEAR FROM fecha_publicacion) = 2023;
```

### Consulta 3
**Nombres de todos los actores que participaron en una película de género comedia.**


**Ejemplo:**
|Nombre       |
|-------------|
|Margot Robbie|
|...          |


**Solución:** Debido a que los datos de los actores que participan en cada pelicula se encuentra en una tabla aparte a la tabla actor y pelicula, primero se combinan las tablas "actor" y "actor_participa_en" por medio de los id de los actores en dichas tablas (actor.id_actor (clave primaria) = actor_participa_en.id_actor (clave foránea)). Luego, a esta unión se le añade la tabla "pelicula" utilizando nuevamente un JOIN por medio de los id de las películas. En este punto se tiene una tabla donde cada fila corresponde a un actor y una película en la que haya participado, siendo id_actor, nombre del actor, id_pelicula, nombre de la película y género de la película las columnas de dicha tabla. Finalmente por medio de un WHERE se filtan las filas de esta tabla de modo que sólo se seleccionen las películas que tengan como género "Comedia". Se usa DISTINCT para que no hayan actores repetidos.

```python
%%sql
SELECT DISTINCT actor.nombre AS "Nombre"
FROM actor
INNER JOIN actor_participa_en_pelicula ON actor.id_actor = actor_participa_en_pelicula.id_actor
INNER JOIN pelicula ON actor_participa_en_pelicula.id_pelicula = pelicula.id_pelicula
WHERE pelicula.genero = 'Comedia';
```

### Consulta 4
**Nombre y cantidad de premios ganados por cada género, ordenados de mayor a menor según la cantidad de premios.**


**Ejemplo:**
|Género  |Cantidad de premios|
|--------|-------------------|
|Drama   |99                 |
|Fantasía|29                 |
|...     |...                |



**Solución:** _Se extrae información sobre los géneros de las películas y la cantidad de premios que ha ganado cada género. Al unir la tabla "pelicula" con "premio_mejor_pelicula" según la correspondencia de sus identificadores de película, se cuenta la cantidad de premios para cada género y se agrupan los resultados por género. Luego, se ordena el resultado de forma descendente según la cantidad de premios_

```python
%%sql
SELECT p.genero AS "Género", COUNT(*) AS "Cantidad de premios" 
FROM pelicula p
INNER JOIN premio_mejor_pelicula pm ON p.id_pelicula = pm.id_pelicula
GROUP BY p.genero
ORDER BY "Cantidad de premios" DESC;
```

### Consulta 5
**Mostrar peliculas que tengan 2 o mas actores/actrices ordenadas de mayor a menor segun la cantidad de actores/actrices.**


**Ejemplo:**
|Nombre     |Cantidad de actores|
|-----------|-------------------|
|Oppenheimer|7                  |
|Poor Things|4                  |
|...        |...                |


**Solución:** Primero se indican las "cosas" a seleccionar en la consulta, que serían nombres de películas y una cantidad contada, la cual apodamos como "Cantidad de actores" (esta última correspondería a la cantidad de actores que hay en cada película seleccionada). Para lograr esta consulta, primero se unieron las tablas "pelicula" y "actor_participa_en_pelicula" por medio de un JOIN utilizando los id de las películas. Luego, se agruparon los resultados por el identificador único de las películas. Esto para separar las películas que tengan el mismo nombre (considerando que esto es posible por la aleatoriedad de los datos de la tabla). Posterior a esto, se aplica una condición con el "HAVING COUNT" para seleccionar sólo las filas que tienen al menos dos actores. Finalmente, se realiza la selección de nombres de películas con su respectiva cantidad de actores y se ordenan los resultados de forma descendente. DESC'.

```python
%%sql
SELECT pelicula.nombre AS "Nombre", COUNT(*) AS "Cantidad de actores"
FROM pelicula
JOIN actor_participa_en_pelicula ON pelicula.id_pelicula = actor_participa_en_pelicula.id_pelicula
GROUP BY pelicula.id_pelicula
HAVING COUNT(*) >= 2
ORDER BY COUNT(*) DESC;
```

### Consulta 6
**Nombre de la película con la mayor brecha de edad entre el actor más joven y el actor más viejo que participen en ella, y las edades de dichos actores.** (Si hay empate, mostrar solamente uno).


**Ejemplo:**
|Nombre|Edad del actor más joven|Edad del actor más viejo|
|------|------------------------|------------------------|
|Barbie|16                      |92                      |


**Solución:** _Se busca la película donde la diferencia de edad entre su actor más joven y su actor más viejo es la mayor. Básicamente, se unen las tablas de películas, actores que participan en esas películas, y los datos de cada actor para calcular quiénes son los más jóvenes y viejos en cada película. Luego, se ordenan estas películas según la brecha de edad (la diferencia entre la edad más vieja y la más joven), y se muestra solo la película con la mayor brecha de edad._

```python
%%sql
SELECT p.nombre AS "Nombre", 
       MIN(a.edad) AS "Edad del actor más joven", 
       MAX(a.edad) AS "Edad del actor más viejo"
FROM pelicula p
INNER JOIN actor_participa_en_pelicula ap ON p.id_pelicula = ap.id_pelicula
INNER JOIN actor a ON ap.id_actor = a.id_actor
GROUP BY p.nombre
ORDER BY MAX(a.edad) - MIN(a.edad) DESC
LIMIT 1;
```

Para terminar con el método a explicar, se utiliza "LIMIT 1" para elegir 1 sola película en caso de que efectivamente existan varias películas con la misma duración más larga. Finalmente, la consulta que acabo de explicar se une a "premio_mejor_pelicula" en base al id de las películas y se selecciona el año de todos los premios de la película de mayor duración seleccionada con la subconsulta.

```python
%%sql
SELECT anio AS "Año"
FROM premio_mejor_pelicula
INNER JOIN (
    SELECT id_pelicula
    FROM pelicula
    WHERE duracion = (
        SELECT MAX(duracion)
        FROM pelicula
    ) LIMIT 1
) AS pelicula_mayor_duracion ON premio_mejor_pelicula.id_pelicula = pelicula_mayor_duracion.id_pelicula;
```

### Consulta 8
**Nombre de todos los actores que han trabajado con más de un director, la cantidad de directores con las que han trabajado y el nombre de una de las películas.**


**Ejemplo:**
|Nombre del actor|Cantidad de directores|Nombre de una película|
|----------------|----------------------|----------------------|
|Cillian Murphy  |29                    |Oppenheimer           |
|...             |...                   |...                   |


**Solución:** _Se dentifican actores que han trabajado con más de un director en diferentes películas. Primero, se crea una tabla temporal llamada "actores_con_mas_de_un_director" que contiene el ID del actor y la cantidad de directores únicos con los que ha trabajado. Luego, se seleccionan el nombre del actor, la cantidad de directores y el nombre de la película en la que han participado estos actores con más de un director. La consulta se une con las tablas "actor", "actor_participa_en_pelicula" y "pelicula" para obtener la información requerida_

```python
%%sql
WITH actores_con_mas_de_un_director AS (
    SELECT ap.id_actor, COUNT(DISTINCT p.id_director) AS cantidad_directores
    FROM actor_participa_en_pelicula ap
    INNER JOIN pelicula p ON ap.id_pelicula = p.id_pelicula
    GROUP BY ap.id_actor
    HAVING COUNT(DISTINCT p.id_director) > 1
)

SELECT a.nombre AS "Nombre del actor", acd.cantidad_directores AS "Cantidad de directores",
       p.nombre AS "Nombre de una película"
FROM actores_con_mas_de_un_director acd
INNER JOIN actor a ON acd.id_actor = a.id_actor
INNER JOIN actor_participa_en_pelicula ap ON a.id_actor = ap.id_actor
INNER JOIN pelicula p ON ap.id_pelicula = p.id_pelicula

```

### Consulta 9

**Nombre de todos los actores que han participado sólo en películas dirigidas por el director con id\_director 1.**


**Ejemplo:**
|Nombre      |
|------------|
|Ryan Gosling|
|...         |


**Solución:** _Escriba aquí la descripción de su solución._

```python
%%sql
SELECT actor.nombre AS "Nombre"
FROM actor
INNER JOIN actor_participa_en_pelicula ON actor.id_actor = actor_participa_en_pelicula.id_actor
INNER JOIN pelicula ON actor_participa_en_pelicula.id_pelicula = pelicula.id_pelicula
WHERE pelicula.id_director = 1
AND actor.id_actor NOT IN (
    SELECT actor_participa_en_pelicula.id_actor
    FROM actor_participa_en_pelicula
    JOIN pelicula ON actor_participa_en_pelicula.id_pelicula = pelicula.id_pelicula
    WHERE pelicula.id_director != 1
);
```

### Consulta BONUS

**Nombre de todos los actores que nunca han ganado un premio, todas las películas en las que han participado y el nombre, número total de películas y número total de premios de los directores de cada película.**


**Ejemplo:**
|Nombre del actor|Nombre de la película|Nombre del director|Número de películas del director|Número de premios del director|
|----------------|---------------------|-------------------|--------------------------------|-----------------------|
|Mark Ruffalo    |Poor Things          |Yorgos Lanthimos   |4                               |0                      |
|Saoirse Ronan   |Lady Bird            |Greta Gerwig       |20                              |1                      |
|Saoirse Ronan   |Little Women         |Greta Gerwig       |20                              |1                      |
|...             |...                  |...                |...                             |...                    |


**Solución:** _Escriba aquí la descripción de su solución._

```python
%%sql
WITH actores_sin_premio AS (
    SELECT DISTINCT a.id_actor, a.nombre
    FROM actor a
    LEFT JOIN premio_mejor_actor pma ON a.id_actor = pma.id_actor
    WHERE pma.id_premio IS NULL
)

SELECT asp.nombre AS "Nombre del actor", p.nombre AS "Nombre de la película",
       d.nombre AS "Nombre del director",
       COUNT(*) OVER (PARTITION BY d.id_director) AS "Numero de peliculas del director",
       SUM(CASE WHEN pm.id_director IS NOT NULL THEN 1 ELSE 0 END) OVER (PARTITION BY d.id_director) AS "Número de premios del director"
FROM actores_sin_premio asp
INNER JOIN actor_participa_en_pelicula ap ON asp.id_actor = ap.id_actor
INNER JOIN pelicula p ON ap.id_pelicula = p.id_pelicula
INNER JOIN director d ON p.id_director = d.id_director
LEFT JOIN premio_mejor_director pm ON pm.id_director = d.id_director
ORDER BY "Nombre del actor", "Nombre de la película", "Nombre del director";
```
